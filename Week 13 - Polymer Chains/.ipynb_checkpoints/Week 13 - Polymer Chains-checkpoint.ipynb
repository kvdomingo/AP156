```python
import numpy as np
import matplotlib.pyplot as mp
import matplotlib.animation as anim
import numpy.random as rd
#from jupyterthemes import jtplot

#jtplot.style()
mp.rcParams["text.usetex"] = True
mp.rcParams["figure.figsize"] = (7,7)
```

```python
class PolymerChain:
    
    def __init__(self, N, maxiters, save=False, savename=None):
        rd.seed(31415)
        self.polymer = np.zeros((N,2), int)
        self.direction = np.zeros((3,2), int)
        for i in range(N):
            self.polymer[i,0] = i - N//2
        self.polymer[:,1] = 0
        self.head = N-1
        self.tail = 0
        self.incr = 1
        self.N = N
        self.maxiters = maxiters
        self.save = save
        self.savename = savename
        
    def choice(self):
        head = self.head
        hm = (self.head - self.incr + self.N)%self.N
        self.direction[0,0] = self.polymer[head,0] - self.polymer[hm,0]
        self.direction[0,1] = self.polymer[head,1] - self.polymer[hm,1]
        self.direction[1,0] = self.direction[0,1]
        self.direction[1,1] = self.direction[0,0]
        self.direction[2,0] = -self.direction[1,0]
        self.direction[2,1] = -self.direction[1,1]
        r = rd.randint(3)
        return self.direction[r]
    
    def intersection(self,c):
        head = self.head
        for i in range(self.N):
            if (c[0] + self.polymer[head,0] == self.polymer[i,0]) and \
            (c[1] + self.polymer[head,1] == self.polymer[i,1]) and \
            (i != self.tail):
                return True
        return False
            
    def accept(self,c):
        head = self.head
        hp = (self.head + self.incr + self.N)%self.N
        self.polymer[hp,0] = c[0] + self.polymer[head,0]
        self.polymer[hp,1] = c[1] + self.polymer[head,1]
        self.head = hp
        self.tail = (self.head + self.incr + self.N)%self.N
        
    def endtoend(self):
        diff = self.polymer[self.head] - self.polymer[self.tail]
        dist = np.sqrt(np.sum(diff**2))
        self.dist = dist
        return dist
        
    def run(self):
        fig = mp.figure(figsize=(10,10), dpi=100, facecolor="k")
        ax = fig.add_subplot(111)
        ax.axis("off")
        ax.grid(False)
        bounds = 4/3*self.N
        ax.set_xlim(-bounds,bounds)
        ax.set_ylim(-bounds,bounds)
        line, = ax.plot(self.polymer[:,0], self.polymer[:,1], "-o", color="white", ms=1, mec="g", mfc="g")
        head, = ax.plot(self.polymer[self.head,0], self.polymer[self.head,1], "ro", ms=8, mfc="None", mew=2)
        tail, = ax.plot(self.polymer[self.tail,0], self.polymer[self.tail,1], "bo", ms=5)
        #ax.relim()
        #ax.autoscale_view(True,True,True)
        data_template = r"\textrm{\\ relative average length $= %.2f$ \\ length of polymer $N = %i$ \\ fractal dimension $= %.2f$}"
        data_text = ax.text(0.025, 0.025, "", transform=ax.transAxes, fontsize=12, color="w")
        
        def init():
            line.set_ydata([np.nan]*len(self.polymer))
            head.set_ydata([np.nan])
            tail.set_ydata([np.nan])
            #ax.relim()
            #ax.autoscale_view(True,True,True)
            data_text.set_text("")
            return line, head, tail, data_text

        def animate(i):
            c = self.choice()
            if self.intersection(c):
                self.head = self.tail
                self.incr *= -1
                self.tail = (self.head + self.incr + self.N)%self.N
            else:
                self.accept(c)
            
            #polysort = self.polymer[self.polymer[:,0].argsort()]
            line.set_xdata(self.polymer[:,0])
            line.set_ydata(self.polymer[:,1])
            head.set_xdata(self.polymer[self.head,0])
            head.set_ydata(self.polymer[self.head,1])
            tail.set_xdata(self.polymer[self.tail,0])
            tail.set_ydata(self.polymer[self.tail,1])
            #ax.relim()
            #ax.autoscale_view(True,True,True)
            data_text.set_text(data_template %(self.endtoend(), self.N, np.log10(self.N)/np.log10(self.endtoend())))
            return line, head, tail, data_text
        
        ani = anim.FuncAnimation(fig, animate, self.maxiters, interval=60, blit=True, init_func=init)
        if self.save:
            ani.save(self.savename, writer="ffmpeg", savefig_kwargs={"facecolor":"black"})
        mp.show()
```

```python
sim = PolymerChain(50,500, True, "polymer50.mp4")
sim.run()
```

```python
sim = PolymerChain(100,500, True, "polymer100.mp4")
sim.run()
```
