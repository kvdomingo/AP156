<b>1.2 The Nonlinear Pendulum</b>

The energy of a nonlinear pendulum is given by:

\begin{equation}
    E = \frac{m}{2}l^2 \dot{\phi}^2 - mgl\cos{\phi} = -mgl\cos{\phi_0}
\end{equation}

where $m$, $g$, $l$ are constants, and $\phi$ is the displacement angle from equilibrium position. Isolating the constants on one side, setting $\kappa := \frac{E}{mgl}$ and $\sqrt{\frac{l}{2g}} := 1$, we have the differential equation

\begin{equation}
    \kappa = \dot{\phi}^2 - \cos{\phi}
\end{equation}

Isolating $\phi$ and $\dot{\phi}$ into separate, coupled differential equations yield

\begin{equation}
    \dot{\phi} = \sqrt{\kappa + \cos{\phi}}
\end{equation}

\begin{equation}
    \phi = \arccos{(\dot{\phi}^2 - \kappa)}
\end{equation}

```python
def T(phi0):
    return 4*sps.ellipk(np.sin(phi0/2)**2)

f,ax = mp.subplots(figsize=(10,10*9/16))
phi0 = np.linspace(0,np.pi,1000)
ax.plot(phi0,T(phi0))
ax.xaxis.set_major_locator(mp.MultipleLocator(np.pi/2))
ax.xaxis.set_minor_locator(mp.MultipleLocator(np.pi/4))
ax.yaxis.set_major_locator(mp.MultipleLocator(2*np.pi))

def format_func_x(value,tick_number):
    N = int(np.round(2*value/np.pi))
    if N == 0:
        return "0"
    elif N == -1:
        return r"$-\pi/2$"
    elif N == 1:
        return r"$\pi/2$"
    elif N == -2:
        return r"$-\pi$"
    elif N == 2:
        return r"$\pi$"
    elif N%2 < 0:
        return r"$-{0}\pi/2$".format(N)
    elif N%2 > 0 or N%2 < 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N//2)
    
def format_func_y(value,tick_number):
    N = int(np.round(value/np.pi))
    if N == 0:
        return "0"
    else:
        return r"${0}\pi$".format(N)
    
ax.xaxis.set_major_formatter(mp.FuncFormatter(format_func_x))
ax.yaxis.set_major_formatter(mp.FuncFormatter(format_func_y))
mp.ylim(0,30)
mp.xlabel("Amplitude $\phi_0$")
mp.ylabel("Period $T$")
mp.title("(1.1) Period as a function of amplitude")
mp.show()
```

```python
def psi(t,phi0):
    return sps.ellipj(t,np.sin(phi0/2)**2)[3]

def sinepsi(t,phi0):
    return sps.ellipj(t,np.sin(phi0/2)**2)[0]

def phinorm(x,phi0):
    return 2*np.arcsin(np.sin(phi0/2) * sinepsi(x*T(phi0),phi0))/phi0

phi0 = [ np.pi/10 , 4*np.pi/5 , 19*np.pi/20 , 99*np.pi/100 , 999*np.pi/1000 ]
x = np.linspace(0,1,100)

def ff(x):
    fflist = []
    for i in range(5):
        fflist.append(phinorm(x,phi0[i]))
    return fflist

flist = ff(x)
```

```python
xx = np.linspace(0,1,100)
xleg = [ "$\pi/10$", "$4\pi/5$", "$19\pi/20$", "$99\pi/100$", "$999\pi/1000$" ]
mp.figure(figsize=(6*16/9,6))
for i in range (5):
    mp.plot(xx,flist[i], label=r"$\phi_0 =$ {0}".format(xleg[i]))
mp.xlabel("$t/T$")
mp.ylabel("$\phi/\phi_0$")
mp.title("(1.2) Normalized displacement $\phi_{norm}(t/T)$")
mp.legend()
#mp.xlim(-1.05,1.05)
mp.show()
```

```python
def dft(y):
    N = len(y)
    b = np.zeros(N, complex)
    s = np.arange(N)
    for r in range(1,N):
        b[r] = sum(y*np.exp(2*np.pi*1j*(s-1)*(r-1)/N))
    return b
```

```python
foulist = np.abs(dft(flist[-1]))/np.sqrt(len(flist[-1]))
mp.figure(figsize=(10,10*9/16))
mp.plot(foulist[:15],"o")
mp.xlabel("$s$")
mp.ylabel("Abs[$b(s)$]")
mp.title("(1.3) Fourier coefficients for $\phi_0 = 0.999\pi$")
mp.show()
```

```python
t = np.linspace(-np.pi,np.pi,200)

def dfi(fi,kappa):
    return np.sqrt(kappa + np.cos(fi))

def fi(dfi,kappa):
    return np.arccos(dfi**2 - kappa)

f,ax = mp.subplots(figsize=(7,7))
ax.set_xlim(-np.pi-0.1,np.pi+0.1)
ax.set_ylim(-np.pi-0.1,np.pi+0.1)
ax.xaxis.set_major_locator(mp.MultipleLocator(np.pi/2))
ax.xaxis.set_major_formatter(mp.FuncFormatter(format_func_x))

for k in [-1/2, 0, 1/2, 1, 3/2, 2]:
    ax.plot(t,dfi(t,k),"b")
    ax.plot(t,-dfi(t,k),"b")
    ax.plot(fi(t,k),t,"b")
    ax.plot(-fi(t,k),t,"b")
    
mp.xlabel("$\phi$")
mp.ylabel("$\dot{\phi}$")
mp.title("(1.4) Phase-space plot of $\dot{\phi}$ vs $\phi$ for different $E$")
mp.show()
```

```python
t = np.linspace(-np.pi,np.pi,250)
fi = np.arccos(t**2)
dfi = np.sqrt(np.cos(fi))

X1,Y1 = np.meshgrid(t, dfi)
Z1 = Y1**2 - np.cos(X1)

X11,Y11 = np.meshgrid(t, -dfi)
Z11 = Y11**2 - np.cos(X11)

X2,Y2 = np.meshgrid(fi,t)
Z2 = Y2**2 - np.cos(X2)

X22,Y22 = np.meshgrid(-fi,t)
Z22 = Y22**2 - np.cos(X22)

fig, ax = mp.subplots(figsize=(7,7))
E = [-1/2, 0, 1/2, 1, 3/2, 2]
ax.contour(X1, Y1, Z1, levels=E,cmap="Blues")
ax.contour(X2, Y2, Z2, levels=E,cmap="Blues")
ax.contour(X11, Y11, Z11, levels=E,cmap="Blues")
ax.contour(X22, Y22, Z22, levels=E,cmap="Blues")

#ax.clabel(CS, inline=1, fontsize=10)
#ax.set_title('Simplest default with labels')
ax.set_xlim(-np.pi-0.1,np.pi+0.1)
ax.set_ylim(-np.pi,np.pi)
ax.xaxis.set_major_locator(mp.MultipleLocator(np.pi/2))
ax.xaxis.set_major_formatter(mp.FuncFormatter(format_func_x))
mp.xlabel("$\phi$")
mp.ylabel("$\dot{\phi}$")
mp.title("(1.4) Phase-space plot of $\dot{\phi}$ vs $\phi$ for different $E$")
mp.show()
```

$3.$ Program the following algorithm for evaluating the complete elliptic integral $K(\sin{\alpha})$ and compare the result with <code>sps.ellipk(np.sin(alpha)**2)</code>

<table width=50%>
    <tr>
        <td>Start:
        <td>$a_0 = 1$
        <td>$b_0 = \cos{\alpha}$
    <tr>
        <td>Iteration:
        <td>$a_{i+1} = \frac{(a_i + b_i)}{2}$
        <td>$b_{i+1} = \sqrt{a_i b_i}$
    <tr>
        <td>Stop:
        <td colspan=2>$|a_n - b_n| < \epsilon$
    <tr>
        <td>Verify:
        <td colspan=2><code>sps.ellipk(np.sin(alpha)**2)</code>$\approx \frac{\pi}{2a_n}$
</table>

The complete elliptic integral of the first kind is defined as

\begin{equation}
    K(k) = \int_0^{\frac{\pi}{2}} \frac{d\theta}{\sqrt{1 - k^2\sin^2{\theta}}} = \int_0^1 \frac{dt}{\sqrt{(1-t^2)(1-k^2 t^2)}}
\end{equation}

So

\begin{equation}
    K(\sin^2{\alpha}) = \int_0^1 \frac{dt}{\sqrt{(1-t^2)(1-t^2\sin^2{\alpha})}}
\end{equation}
