```python
import numpy as np
import matplotlib.pyplot as mp
import matplotlib.animation as anim
import numpy.random as rd

%matplotlib qt5
mp.rcParams["text.usetex"] = True
mp.rcParams["figure.figsize"] = (7,7)
```

```python
class Percolation(object):

    def __init__(self,n,p=0.59275):
        rd.random(314)
        self.p = p
        self.n = n
        a = np.zeros((n,n), int)
        self.a = a
        label = np.zeros_like(self.a, int)
        label[:,0] = a[:,0]*-1
        label[0,:] = a[0,:]*-1
        label[:,n-1] = a[:,n-1]*-1
        label[n-1,:] = a[n-1,:]*-1
        label[n//2,n//2] = 1
        self.label = label
        self.clustersize = None

    def mass(self,label): 
        elements, counts = np.unique(label, return_counts=1)
        return np.mean(counts[elements!=0])
        
    #Hoshen-Kopelman algorithm
    def labelling(self): #Create the clusters and label them
        a = self.a
        label = self.label
        n = self.n

        for i in range(1,n-1):
            for j in range(1,n-1):
                if a[i,j]: #if occupied
                    if a[i-1, j] and not(a[i,j-1]): #if up is occupied and left is not occupied
                        label[i,j] = label[i-1,j] #move up
                    elif a[i, j-1] and not(a[i-1, j]): #if left is occupied and up is not occupied
                        label[i,j] = label[i, j-1] #move left
                    elif not(a[i,j-1]) and not(a[i-1, j]):
                        label[i,j] = np.max(label) + 1
                    elif a[i-1, j] and a[i, j-1]:
                        if label[i-1,j] < label[i, j-1]:
                            label[i,j] = label[i-1,j]
                            label[label == label[i, j-1]] = label[i-1,j]
                        elif label[i-1,j] > label[i, j-1]:
                            label[i,j] = label[i,j-1]
                            label[label == label[i-1, j]] = label[i,j-1]
                        else:
                            label[i,j] = label[i, j-1] 
                    if i == n-2:
                        if label[i+1, j] == -1:
                            label[label == label[i,j]] = -1
                    if j == n-2:
                        if label[i, j+1] == -1:
                            label[label == label[i,j]] = -1
                            
        label[label==-1] = 0
        clustersize = self.mass(label)
        self.label = label
        self.a = a
        self.label = label
        self.clustersize = clustersize
    
    def run_animate(self):
        fig = mp.figure(figsize=(7,7))
        ax = fig.add_subplot(111)
        ax.grid(False)
        
        def init():
            xf = mp.imshow(self.label, cmap="gray", animated=True)
            return xf,
        
        def animate(i):
            self.labelling()
            xf = mp.imshow(self.label, cmap="gray", animated=True)
            return xf,
        
        ani = anim.FuncAnimation(fig, animate, 50, interval=30, blit=True, init_func=init)
        ani.save("percolation.mp4", writer="ffmpeg")
        mp.show()
        
    #percolation threshold
    def thresh(self,n = 50,trials = 10):
        pl = np.arange(0, 1.1, 0.01)
        massl = np.empty_like(pl)
        for w in range(0, 100):
            p = pl[w]
            massp = []
            for _ in range(trials):
                print(' trial', len(massp))
                print(' p= ', p)
                __, __, clustersize = self.labelling(n,p)
                massp.append(clustersize)
            if len(massp) != 0:
                massl[w] = np.mean(massp)
            else:
                massl[w] = 0
        return pl, massl
    
    def labellingplot(self):
        a, label, _ = self.labelling(500,0.59275)
        mp.figure(figsize=(7,7))
        mp.imshow(a, cmap = 'gray')
        mp.title(r"\textrm{Percolation at p = %s}"%(0.59275))
        mp.savefig('Percolation.png', dpi=300, bbox_inches='tight')

        mp.figure(figsize=(7,7))
        mp.imshow(label)
        mp.title(r"\textrm{Size clusters at p = %s}"%(0.59275))
        mp.savefig('Labelling.png', dpi=300, bbox_inches='tight')


    def Percothreshplot(self):
        pl,massl = self.thresh(n = 500, trials = 10)
        pthreshold = pl[massl == np.max(massl)]
        print('p_c = ', pthreshold )
        mp.figure()        
        mp.plot(pl, massl,'--ro', lw = 2, ms = 6)
        mp.xlabel("$p$", fontsize = 12) #Concentration
        mp.ylabel("$M$", fontsize = 12) #Cluster Mass   
        mp.savefig('Mvsp.png', dpi = 300, bbox_inches = 'tight')
```

```python
sim = Percolation(500)
sim.run_animate()
```

```python
sim = Percolation()
sim.labellingplot()
sim.Percothreshplot()
```

```python
class perco2:
    
    def __init__(self, L):
        rd.seed(100)
        self.L = L
        self.lattice = np.zeros((L,L), int)
        self.pr = rd.random()
        self.p = 0.59275
        for i in range(L):
            for j in range(L):
                if rd.random() < self.p:
                    self.lattice[i,j] = 1
                    
    def show_percolation(self, save=False):
        mp.figure()
        mp.title(r"\textrm{(5.6) Percolation structure at $p_c$}")
        mp.imshow(self.lattice, cmap="gray")
        if save:
            mp.savefig("percolation.png", dpi=300, bbox_inches="tight")
        mp.show()
        
    def labelling(self):
        #define NOTDEFINED 0
        #define OCCUPIED 1
        #define VACANT 2
        L = self.L
        arr = np.zeros((L,L), int)
        arr[0,:] = 2
        arr[:,0] = 2
        arr[:,L-1] = 2
        arr[L-1,:] = 2
        arr[L//2, L//2] = 1
        self.pd = 4*L
        li = np.zeros((self.pd,2), int)
        li[0] = [L//2+1, L//2]
        li[1] = [L//2, L//2+1]
        li[2] = [L//2-1, L//2]
        li[3] = [L//2, L//2-1]
        self.count = 3
        self.li = li
        self.arr = arr
        
    def definition(i,j):
        count = self.count
        li = self.li
        arr = self.arr
        pd = self.pd
        
        if arr[i,j] != 0:
            self.count = count
            self.arr = arr
            self.li = li
            return
        
        if rd.random() < self.p:
            arr[i,j] = 1
            if arr[i,j+1] == 0:
                count += 1
                li[count%pd,0], li[count%pd,1] = i, j+1
            if arr[i,j-1] == 0:
                count += 1
                li[count%pd,0], li[count%pd,1] = i, j-1
            if arr[i+1,j] == 0:
                count += 1
                li[count%pd,0], li[count%pd,1] = i+1, j
            if arr[i-1,j] == 0:
                count += 1
                li[count%pd,0], li[count%pd,1] = i-1, j
            else:
                arr[i,j] = 2

        self.count = count
        self.arr = arr
        self.li = li
        
        while True:
            for k in range(self.count+1):
                i,j = li[k%pd,0], li[k%pd,1]
                self.definition(i,j)
                
    #def run_anim(self):
     #   fig = mp.figure(figsize=(7,7))
      #  ax = fig.add_subplot(111)
       # li = self.li
        #pd = self.pd
        
      #  def init():
       #     xf = mp.imshow(self.arr, cmap="gray", animated=True)
        #    return xf,
        
#            for k in range(self.count+1):
 #               i,j = li[k%pd,0], li[k%pd,1]
  #              self.definition(i,j)
   #         xf = mp.imshow(self.arr, cmap="gray", animated=True)
    #        return xf,
     #   
      #  ani = anim.FuncAnimation(fig, animate, 100, interval=30, blit=True, init_func=init)
      #  ani.save("percolation.mp4", writer="ffmpeg")
      #  mp.show()
                    
    def show_cluster(self, save=False):
        mp.figure()
        mp.title(r"\textrm{(5.7) Percolating cluster at $p_c$}")
        mp.imshow(self.arr, cmap="gray")
        if save:
            mp.savefig("labelling.png", dpi=300, bbox_inches="tight")
        mp.show()
```

```python
sim = perco2(500)
sim.show_percolation()
sim.labelling()
sim.show_cluster()
```
